###  K-Means cluster algorithm

```python
from sklearn.cluster import KMeans

# Removing the labels column
x = data.drop(['label'], axis = 1)

# select all the remaining values of the data set
x = x.values

print(x.shape)
```

```python
# lets determine the optimum number of clusters in dataset

plt.rcParams['figure.figsize'] = (10, 4)
wcss = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, init= 'k-means++', max_iter = 300, n_init= 10, random_state=0)
    km.fit(x)
    wcss.append(km.inertia_)
    
# lets plot the results
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method', fontsize = 20)
plt.xlabel('No. of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
# lets implement the K Means algorithm to perform clustring analysis

km = KMeans(n_clusters= 4, init = 'k-means++', max_iter = 300, n_init= 10, random_state = 0)
y_means = km.fit_predict(x)

# find out the results
a = data['label']
y_means = pd.DataFrame(y_means)
z = pd.concat([y_means, a], axis=1)
z = z.rename(columns= { 0:'cluster'})

# check clusters of each crops
print('check the results after applying the k Means clustering analysis \n')
print('Crops in First cluster:', z[z ['cluster'] == 0]['label'].unique())
print('--------------------------------------------------')
print('Crops in second cluster:', z[z ['cluster'] == 1]['label'].unique())
print('------------------------------------------------')
print('Crops in Third cluster:', z[z ['cluster'] == 2]['label'].unique())
print('--------------------------------------------------')
print('Crops in Fourth cluster:', z[z ['cluster'] == 3]['label'].unique())
```

```python
# split the Dataset for making predictive modeling

y = data['label']
x = data.drop(['label'], axis = 1)

print('x:', x.shape)
print('y:', y.shape)
```

```python
# Train test spliting

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state=0)
print('x train shape:', x_train.shape)
print('x test shape:', x_test.shape)
print('y train shape:', y_train.shape)
print('y test shape:', y_test.shape)
```

###   Logistic Regression

```python
# create the predictive Model

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lr_model = LogisticRegression()
lr_model.fit(x_train, y_train)
lr_y_pred = lr_model.predict(x_test)
```

```python
accuracy_score(y_test, lr_y_pred)
```

## Confusion matrix, Accuracy  Precision, Recall and F1-score

    precision :
            In pattern recognition, information retrieval and binary classification, precision (also called positive predictive value) is the fraction of relevant instances among the retrieved instances,

    recall :
        recall is the fraction of relevant instances that have been retrieved over the total amount of relevant instances.

    F-score :
        the F1 score is a measure of a test's accuracy. It considers both the precision p and the recall r of the test to compute the score: p is the number of correct positive results divided by the number of all positive results returned by the classifier, and r is the number of correct positive results divided by the number of all relevant samples (all samples that should have been identified as positive). The F1 score is the harmonic average of the precision and recall, where an F1 score reaches its best value at 1 (perfect precision and recall) and worst at 0.

```python
# lets check the model performance
from sklearn.metrics import confusion_matrix, classification_report

# print the confusion matrics
plt.rcParams['figure.figsize'] = (10, 10)
cm = confusion_matrix(y_test, lr_y_pred)
sns.heatmap(cm, annot = True, cmap = 'Wistia')
plt.title('Confusion Matrix for Logistic Regression', fontsize = 15)
plt.show()
```

```python
# lets print classification Report
cr = classification_report(y_test, lr_y_pred)
print(cr)
```

### Support vector machine algorithm

```python
# Support Vector Machine
from sklearn.svm import SVC

sv_model = SVC()
sv_model.fit(x_train, y_train)

sv_y_pred = sv_model.predict(x_test)

# Summary of the predictions made by the classifier
print(classification_report(y_test, sv_y_pred))
print(confusion_matrix(y_test, sv_y_pred))
# Accuracy score

print('accuracy is',accuracy_score(sv_y_pred,y_test))
```

### Decision Tree model

```python
# Decision Tree's
from sklearn.tree import DecisionTreeClassifier

dt_model= DecisionTreeClassifier()

dt_model.fit(x_train, y_train)

dt_y_pred = dt_model.predict(x_test)

# Summary of the predictions made by the classifier
print(classification_report(y_test, dt_y_pred))
print(confusion_matrix(y_test, dt_y_pred))
# Accuracy score
print('accuracy is',accuracy_score(dt_y_pred,y_test))
```

### Random Forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier(max_depth=2)
rf_model.fit(x_train,y_train)
rf_y_pred=rf_model.predict(x_test)
print(classification_report(y_test,rf_y_pred))
print(confusion_matrix(rf_y_pred,y_test))
#Accuracy Score
print('accuracy is ',accuracy_score(rf_y_pred,y_test))
```
